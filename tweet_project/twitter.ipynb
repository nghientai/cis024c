```python
import tweetClient

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

def main():
    
    # Alow input of tweet key from user    
    tweetKey = raw_input("Enter a Tweet keywork (Ex:NFL,Python..): ")
    numberOfPages = raw_input("How many items do you want to get? (Max 500): ")
    print("******************************************************************************************\n\n")   
    # creating object of TwitterClient Class
    api = tweetClient.TwitterClient()
    # calling function to get tweets
    tweets = api.get_tweets(query = tweetKey, count = numberOfPages) #the maximum count is 500
 
    # picking positive tweets from tweets
    ptweets = [tweet for tweet in tweets if tweet['sentiment'] == 'positive']
    ntweets = [tweet for tweet in tweets if tweet['sentiment'] == 'negative']
    lenghOFThree = len(tweets) - len(ntweets) - len(ptweets)
    # percentage of positive tweets
    positive = format(100*len(ptweets)/len(tweets))
    negative = format(100*len(ntweets)/len(tweets))
    neutral = format((100*lenghOFThree)/len(tweets))
   
    # Add 3 values above to a list and call sentement() below
    listSentement = [positive,negative,neutral]
    sentement(listSentement)

    # printing first 5 positive tweets
    print("******************************************************************************************")
    print("*********************************** POSITIVE TWEETS **************************************")
    print("******************************************************************************************")
    for tweet in ptweets[:5]:
        print("+ "+tweet['text'])
    # printing first 5 negative tweets
    print("\n\n")
    print("******************************************************************************************")
    print("*********************************** NEGATIVE TWEETS **************************************")
    print("******************************************************************************************")
                
    for tweet in ntweets[:5]:
        print("+ "+tweet['text'])
    

def sentement(listSentement):
    
    objects = ('Positive', 'Negative', 'Neutral')
    y_pos = np.arange(len(objects))
    performance = listSentement

    plt.title('Sentiment analysis')        
    plt.pie(performance, 
            labels = objects,
            autopct = '%1.1f%%',
            startangle = 90)
        
    plt.show()
        
if __name__ == "__main__":
    # calling main function
    main()
```
